```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6] 
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
speed1 = [85,99,111,86,97,94,87,78,11,55,99,77,15]
```

```python
plt.scatter(age, speed, marker = 's', color = 'Olive')
plt.scatter(age, speed1, marker = 'h', color = 'Purple')
plt.xlabel("Age",fontsize = 15)
plt.ylabel("Speed", fontsize = 20)
plt.title("Car comparision", fontsize = 25)
plt.legend(['speed', 'speed1'], loc = "lower right", ncol = 2, facecolor = "yellow", fontsize = 15)
plt.show()
```
